```python
# Câu 5:
# 5.1: Khai báo biến:
# Cú pháp:
# tên_biến = giá_trị
# Ví dụ:
x = 10  # Biến x có kiểu dữ liệu int (số nguyên)
name = "John"  # Biến name có kiểu dữ liệu str (chuỗi)
is_active = True  # Biến is_active có kiểu dữ liệu bool (giá trị logic)

# 5.2: Các toán tử:
# Toán tử số học: +, -, *, /, //, %, **
# Toán tử so sánh: ==, !=, >, <, >=, <=
# Toán tử logic: and, or, not
#Toán tử gán: =, +=, -=, =, /=, //=, %=, *=
# Ví dụ:
a = 10
b = 5
# Toán tử số học
print(a + b)  # Kết quả: 15
print(a - b)  # Kết quả: 5
print(a * b)  # Kết quả: 50
print(a / b)  # Kết quả: 2.0
print(a // b)  # Kết quả: 2
print(a % b)  # Kết quả: 0
print(a ** b)  # Kết quả: 100000
# Toán tử so sánh
print(a == b)  # Kết quả: False
print(a != b)  # Kết quả: True
print(a >= b)  # Kết quả: True
# Toán tử logic
print(a >= b and b >= 0)  # Kết quả: True
print(a <= b or b >= 0)  # Kết quả: True
print(not a <= b)  # Kết quả: True
# Toán tử gán
a += 5  # Tương đương a = a + 5
print(a)  # Kết quả: 15

# 5.3: Các kiểu dữ liệu

# Python hỗ trợ nhiều kiểu dữ liệu khác nhau, bao gồm:

# Số nguyên (int): 1, 2, 3, -1, -2, -3, ...

# #Số thực (float): 1.0, 2.5, -3.14, ...

# Chuỗi (str): "Hello", "Python", "123", ...

# Giá trị logic (bool): True, False

# Danh sách (list): [1, 2, 3], ["apple", "banana", "cherry"], ...

# Bộ (tuple): (1, 2, 3), ("apple", "banana", "cherry"), ...

# Từ điển (dict): {"name": "John", "age": 30}, ...
# Ví dụ:

# Số nguyên
number = 10
print(number)  # Kết quả: 10
#Số thực
pi = 3.14
print(pi)  # Kết quả: 3.14
# Chuỗi
message = "Hello, world!"
print(message)  # Kết quả: Hello, world!
# Giá trị logic
is_active = True
print(is_active)  # Kết quả: True
# Danh sách
fruits = ["apple", "banana", "cherry"]
print(fruits)  # Kết quả: ['apple', 'banana', 'cherry']
# Bộ
coordinates = (1, 2, 3)
print(coordinates)  # Kết quả: (1, 2, 3)
# Từ điển
person = {"name": "John", "age": 30}
print(person)  # Kết quả: {'name': 'John', 'age': 30}

```
